```python
# df: dataframe
# column: which column to expand
#timestamp: which column is the timestamp
# lag_width: the windows size

def shift_expand(df, column, timestamp, lag_width):
    temps = DataFrame(temp_df[column].values)
    #lag_width = 10  #number of lagging elements, concat with current time, total 4
    temps_list = list()
    
    for i in range(0,lag_width+1):
        temps_list.append(temps.shift(i))
        
    dataframe = DataFrame()
    
    for i in range(len(temps_list)-1,-1,-1):
        dataframe = concat([dataframe, temps_list[i]], axis=1)
        
    rename_list = list()
    
    for i in range(0,len(temps_list)):
        rename_list.append('t-'+ str(len(temps_list)-1-i)+'_'+column)
        
    dataframe.columns = rename_list
    dataframe['timestamp (SGT)'] = df[timestamp].reset_index(drop=True) # must reset index
    #dataframe['timestamp (SGT)'] = temp_df
    
    return dataframe    
```
